```python
import pandas as pd
import numpy as np
import google.generativeai as genai

# Pega a chave do secrets no colab
from google.colab import userdata
GOOGLE_API_KEY = userdata.get('key_genai')
#print(GOOGLE_API_KEY)

genai.configure(api_key=GOOGLE_API_KEY)
```

```python
# modelos suportados
for m in genai.list_models():
  if 'embedContent' in m.supported_generation_methods:
    print(m.name)
```

```python
# Teste de geração de embeddings
model = "models/embedding-001"
text = "Hello World"
result = genai.embed_content(model=model,content=text)
print(result['embedding'])
```

```python
title = "A proxima geração de ai para devs"
sample_text = ("Tiulo: A proxima geração de ai para devs"
"\n"
"Artigo completo:\n"
"\n"
"Gemini e Google ")
emb = genai.embed_content(model="models/embedding-001",content=sample_text,title=title,task_type="RETRIEVAL_DOCUMENT")
print(emb)
```

```python
# Juntando os documentos
DOCUMENT1 = {
    "Título": "Mudanca de marchas",
    "Conteúdo":"Seu google car tem uma transmissao automatica. Para trocar marchar mova a alavanca."
}

DOCUMENT2 = {
    "Título": "Touchscreen",
    "Conteúdo":"Basta passar o dedo para acessar as funcionalidades."
}

DOCUMENT3 = {
    "Título": "sistema operacional",
    "Conteúdo":"Sistema operacional de ultima geracao, tornando mais rapido."
}

documents = [DOCUMENT1,DOCUMENT2,DOCUMENT3]

```

```python
df = pd.DataFrame(documents)
df.columns = ["Titulo","Conteudo"]
df
```

```python
model = "models/embedding-001"
```

```python
def emb_fn(title,text):
  return genai.embed_content(
      model=model,
      content=sample_text,
      title=title,
      task_type="RETRIEVAL_DOCUMENT")["embedding"]
```

```python
df["Embeddings"] = df.apply(lambda row: emb_fn(row["Titulo"],row["Conteudo"]), axis=1 )
df
```

```python
def gerar_e_buscar_consulta(consulta,base, model):
  emb_query= genai.embed_content(
      model=model,
      content=consulta,
      task_type="RETRIEVAL_QUERY")["embedding"]
  produtos_escalares = np.dot(np.stack(df["Embeddings"]),emb_query)
  indice = np.argmax(produtos_escalares)
  return df.iloc[indice]["Conteudo"]
```

```python
consulta = "Como consigo passar marcha?"
txt = gerar_e_buscar_consulta(consulta,df,model)
print(txt)
```

```python
prompt = f"Reescreva o texto em italiano, sem adicionar informações que não façam parte do texto: {txt}"
model_gen = genai.GenerativeModel("gemini-1.0-pro")
response = model_gen.generate_content(prompt)
print(response.text)
```
